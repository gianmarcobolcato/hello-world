# Codiv-19 Italy update 18/03/2020 

```python
import pandas as pd
df=pd.read_csv('ita.csv',sep=',')
```

Cumulative Tren for NewCases and Deaths

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots(figsize=(12,8));
plt.title('Cumulative Trend 18/03/2020')
sns.set(style="ticks", rc={"lines.linewidth": 2.5})
ax.plot(df['data'],df['totale_casi'],c='blue')
plt.xlabel('date')
plt.ylabel('total_cases')
plt.xticks(df['data'], rotation='vertical')
ax2=ax.twinx()
ax2.plot(df['data'],df['deceduti'],c='red')
ax2.set_ylabel('total_deaths')
plt.show()
```
