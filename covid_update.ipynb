# Covid-19 ITALY/CHINA update 18/03/2020

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
df18=pd.read_excel('COVID-19-geographic-disbtribution-worldwide-2020-03-17.xlsx')

country=['Italy','China']
df18=df18.loc[(df18.CountryExp.isin(country)),:]

fig, ax = plt.subplots(figsize=(15,8));
plt.title('Trend update 18/03/2020')
sns.set(style="ticks", rc={"lines.linewidth": 2.5})
g=sns.lineplot(data=df18, x='DateRep',y='NewConfCases',hue='CountryExp');
plt.legend(loc='upper left', labels=['Cases in China ', 'Cases in Italy'])
ax2=ax.twinx()
g=sns.lineplot(data=df18, x='DateRep',y='NewDeaths',hue='CountryExp',ax=ax2,dashes=True);
for i in range(len(country)):
    g.lines[i].set_linestyle("--")
    plt.legend(loc='lower left', labels=['Death in China ', 'Death in Italy'])
plt.show()
```

left y-axis is for NewCase, right y-axis is for NewDeath

```python
d18=df18.groupby('CountryExp')['NewConfCases','NewDeaths'].sum().reset_index()
d18
```

Total cases and deaths

```python
for i in d18['CountryExp']:
    dd18=d18[d18['CountryExp']==i]
    death_rate=set((dd18['NewDeaths']/dd18['NewConfCases'])*100)
    print('the death rate in {} is={} %'.format(i,death_rate));
```
